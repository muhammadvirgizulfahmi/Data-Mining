```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('TVShow.csv', header=None)
dataset.head()
```

```python
shows = dataset.values.astype(str).tolist()
```

```python
from apyori import apriori

rules = apriori(transactions=shows,min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
result = list(rules)
result
```

```python
def inspect(result):
    lhs = [tuple(row[2][0][0])[0] for row in result]
    rhs = [tuple(row[2][0][1])[0] for row in result]
    support = [row[1] for row in result]
    confidence = [row[2][0][2] for row in result]
    lift = [row[2][0][-1] for row in result]
    return list(zip(lhs,rhs,support,confidence,lift))
resultdf = pd.DataFrame(inspect(result), columns=['LHS','RHS','Support','Confidence','Lift'])
```

```python
resultdf.sort_values(by=['Lift'],axis=0,ascending=False)
```
