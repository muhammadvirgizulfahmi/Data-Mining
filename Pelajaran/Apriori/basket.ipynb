```python
import pandas as pd
```

```python
# Memuat dataset
data = pd.read_csv('basket.csv', header=None)

# Menampilkan beberapa baris pertama
print(data.head())
```

```python
# Mengubah data menjadi format transaksi
transactions = data.stack().groupby(level=0).apply(list).tolist()
transactions
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Mengonversi transaksi ke dalam format one-hot encoding
encoder = TransactionEncoder()
onehot = encoder.fit(transactions).transform(transactions)
df = pd.DataFrame(onehot, columns=encoder.columns_)

# Mendapatkan itemset yang sering muncul dengan minimum support 0.01
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
# rules = apriori(transactions=shows,min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
# Menemukan aturan asosiasi
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
```
