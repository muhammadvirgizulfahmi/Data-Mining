```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read data
data = pd.read_csv("data_BMI.csv")

# cetak/tampilkan data
print(data)
```

```python
# membuat variabel berat badan wanita
female_weight = data.loc[data['Gender'] == 'Female', 'Weight']

# menghitung jumlah data berat badan wanita
print("Jumlah data berat badan wanita: ")
print(female_weight.count())

# menghitung rata-rata berat badan wanita
print("\nRata-rata berat badan wanita: ")
print(female_weight.mean())

# menghitung nilai maksimum berat badan wanita
print("\nNilai maksimum berat badan wanita: ")
print(female_weight.max())

# menghitung nilai minimum berat badan wanita
print("\nNilai minimum berat badan wanita: ")
print(female_weight.min())
```

```python
# membuat variabel tinggi wanita
female_height = data.loc[data['Gender'] == 'Female', 'Height']

# menghitung jumlah data tinggi wanita
print("Jumlah data tinggi wanita: ")
print(female_height.count())

# menghitung rata-rata tinggi wanita
print("\nRata-rata tinggi wanita: ")
print(female_height.mean())

# menghitung nilai maksimum tinggi wanita
print("\nNilai maksimum tinggi wanita: ")
print(female_height.max())

# menghitung nilai minimum tinggi wanita
print("\nNilai minimum tinggi wanita: ")
print(female_height.min())
```

```python
# membuat variabel berat badan pria
male_weight = data.loc[data['Gender'] == 'Male', 'Weight']

# menghitung jumlah data berat badan pria
print("Jumlah data berat badan pria: ")
print(male_weight.count())

# menghitung rata-rata berat badan pria
print("\nRata-rata berat badan pria: ")
print(male_weight.mean())

# menghitung nilai maksimum berat badan pria
print("\nNilai maksimum berat badan pria: ")
print(male_weight.max())

# menghitung nilai minimum berat badan pria
print("\nNilai minimum berat badan pria: ")
print(male_weight.min())
```

```python
# membuat variabel tinggi pria
male_height = data.loc[data['Gender'] == 'Male', 'Height']

# menghitung jumlah data tinggi pria
print("Jumlah data tinggi pria: ")
print(male_height.count())

# menghitung rata-rata tinggi pria
print("\nRata-rata tinggi pria: ")
print(male_height.mean())

# menghitung nilai maksimum tinggi pria
print("\nNilai maksimum tinggi pria: ")
print(male_height.max())

# menghitung nilai minimum tinggi pria
print("\nNilai minimum tinggi pria: ")
print(male_height.min())
```

```python
# mengubah data gender; Female menjadi 0, Male menjadi 1
data.loc[data['Gender'] == 'Female', 'Gender'] = 0
data.loc[data['Gender'] == 'Male', 'Gender'] = 1

print(data)
```

```python
plt.figure(figsize=(8,7))
sns.countplot(x='Index', hue='Gender', data=data, palette='Set1')
plt.xticks([0, 1, 2, 3, 4, 5],
           ['Very Weak', 'Weak', 'Normal', 'Overweight', 'Obesity', 'Extreme Obesity'])
plt.title('Distribusi Index BMI berdasarkan Gender')
plt.xlabel('Index BMI')
plt.ylabel('Jumlah')
plt.show()
```

```python
sns.displot(data=data, x="Index", hue="Gender", multiple='dodge',  height=3, aspect=3)
plt.xticks([0, 1, 2, 3, 4, 5], ['Very Weak', 'Weak', 'Normal', 'Overweight', 'Obesity', 'Extreme Obesity'])
plt.title('Distribusi Index BMI berdasarkan Gender')
plt.xlabel('Index BMI')
plt.ylabel('Gender')
plt.show()
```

```python
sns.histplot(data=data, x="Index", hue="Gender", multiple='dodge', shrink=1)
plt.xticks([0, 1, 2, 3, 4, 5], ['Very Weak', 'Weak', 'Normal', 'Overweight', 'Obesity', 'Extreme Obesity'])
plt.title('Distribusi Index BMI berdasarkan Gender')
plt.xlabel('Index BMI')
plt.ylabel('Gender')
plt.show()
```
