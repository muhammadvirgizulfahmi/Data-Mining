```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('BL-Flickr-Images-Book.csv')
df
```

```python
# melihat datatype dan spesifik pada dataset
df.info()
```

```python
# mengecek apakah ada missing value
df.isnull().sum()
```

```python
# dropping unnecessary columns atau
# hapus data yang tidak diperlukan karna keseluruhan data memiliki nilai kosong
columns_to_drop = ['Edition Statement', 'Corporate Author',
                    'Corporate Contributors', 'Former owner', 'Engraver']
df.drop(columns=columns_to_drop, inplace=True)
df
```

```python
# menjadikan identifier sebagai index untuk memudahkan mengambil sebuah data
df.set_index('Identifier', inplace=True)
df
```

```python
# renaming columns
df.rename(columns={
    'Place of Publication':'Tempat_Penerbitan',
    'Date of Publication':'Tahun_Penerbitan',
    'Publisher':'Penerbit',
    'Title':'Judul',
    'Author':'Penulis',
    'Contributors':'Kontributor',
    'Issuance type':'Jenis_Penerbitan',
    'Flickr URL':'URL_Flickr',
    'Shelfmarks':'Penanda_Rak'
}, inplace=True)

# tampilkan dataframe setelah pengubahan nama kolom
df
```
